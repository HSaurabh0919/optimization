```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_geometric_decay(
    initial_value: float, 
    decay_rate: float, 
    time_periods: int
) -> np.ndarray:
    """
    Calculates a series using geometric decay.

    Args:
        initial_value (float): The starting value at time t=0.
        decay_rate (float): The rate at which the value decays per period 
                            (a value between 0.0 and 1.0).
        time_periods (int): The number of time steps (t=0, 1, 2, ..., N-1).

    Returns:
        np.ndarray: An array of values showing the geometric decay over time.
    """
    if not 0.0 <= decay_rate <= 1.0:
        raise ValueError("Decay rate must be between 0.0 and 1.0.")
    if time_periods <= 0:
        raise ValueError("Time periods must be a positive integer.")

    # Create an array of time steps (t=0, 1, 2, ...)
    t = np.arange(time_periods)
    
    # Apply the geometric decay formula using NumPy's power function
    decay_series = initial_value * (decay_rate ** t)/decay_rate
    
    return decay_series


```

```python

# 1. Define Parameters
initial_amount = 100.0   # The starting value (e.g., initial impact)
decay_factor_A = 0.8     # A slow decay (80% carries over each period)
decay_factor_B = 0.4     # A fast decay (40% carries over each period)
num_steps = 20           # Calculate for 10 time periods

x = [i for i in range(0,num_steps)]
decay_factors_list = [i/10 for i in range(1,10)]
decay_value_list = [calculate_geometric_decay(
    initial_amount, decay_factor, num_steps
) for decay_factor in decay_factors_list]
```

```python
# plot
plt.figure(figsize=(8, 5))
for alpha, values in zip(decay_factors_list, decay_value_list):
    plt.plot(x, values, label=f"alpha={alpha}")

plt.title("Reach Value Curves for Different Alpha Values")
plt.xlabel("x")
plt.ylabel("Reach Value")
plt.legend()
plt.grid(True)
plt.show()
```
