```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# your function
def compute_reach_value(x, mu):
    x = np.array(x, dtype=float)        # convert list to NumPy array
    z = -x / mu if mu != 0 else -x      # handle mu = 0 separately
    z = np.clip(z, -500, 500)           # avoid overflow
    f = (1 - np.exp(z)) / (1 + np.exp(z))
    return f
```

```python
mu_list = [i for i in range(0,9)]
x = [i for i in range(0,11)]
value_list = [compute_reach_value(x,mu) for mu in mu_list]
```

```python
# compute values for each mu
value_list = [compute_reach_value(x, mu) for mu in mu_list]

# plot
plt.figure(figsize=(8, 5))
for mu, values in zip(mu_list, value_list):
    plt.plot(x, values, label=f"mu={mu}")

plt.title("Reach Value Curves for Different Mu Values")
plt.xlabel("x")
plt.ylabel("Reach Value")
plt.legend()
plt.grid(True)
plt.show()
```
